# Задание по генерации структуры водорода

```python
import numpy as np
```

```python
def get_h2(r):
    alpha=np.random.rand()*np.pi
    betta=np.random.rand()*2*np.pi
    data=[]
    h=0.37
    z_h=h*np.sin(alpha)
    x_h=h*np.cos(alpha)*np.cos(betta)
    y_h=h*np.cos(alpha)*np.sin(betta)
    r_h=np.array([x_h,y_h,z_h])
    data.append([0, r+r_h])
    data.append([0, r-r_h])
    h=0.1
    z_h=h*np.sin(alpha)
    x_h=h*np.cos(alpha)*np.cos(betta)
    y_h=h*np.cos(alpha)*np.sin(betta)
    r_h=np.array([x_h,y_h,z_h])
    data.append([1, r+r_h])
    h=-0.2
    z_h=h*np.sin(alpha)
    x_h=h*np.cos(alpha)*np.cos(betta)
    y_h=h*np.cos(alpha)*np.sin(betta)
    r_h=np.array([x_h,y_h,z_h])
    data.append([-1, r+r_h])

    return data

def get_structure(a=17.7, N=500):
    data=[]
    for x_i in np.linspace(0,a,10):
        for y_i in np.linspace(0,a,10):
            for z_i in np.linspace(0,a/2,5):
                random_flag=np.random.rand()
                if (random_flag > 0.5):
                    data += get_h2(np.array([x_i,y_i,z_i]))
                else:
                    data += get_h2(np.array([x_i,y_i,a-z_i]))
                    
    return data

def write_lammps_data(filename, a, N_h2):
    N_particles=N_h2*4
    print ("Generate Structure for box={:f} and {:d} H2".format(a,N_h2))
    data=get_structure(a=a,N=N_h2)
    
    with open(filename,"w") as f:
        f.write("Created by Ilya Fedorov for Hydrogen research\n"+
                "\n"+
                "{:d} atoms\n".format(N_particles)+
                "2 atom types\n"
                "\n"+
                "0.000000 {:.6f} xlo xhi\n".format(a)+
                "0.000000 {:.6f} ylo yhi\n".format(a)+
                "0.000000 {:.6f} zlo zhi\n".format(a)+
                "\n"+
                "Masses\n"+
                "\n"+
                "1 1.007940\n"+
                "1 1.000000\n"+
                "\n"+
                "Atoms\n"+
                "\n")
        print ("Write coordinates")
        for i in range(N_particles):
            part=data[i]
            if part[0]==0:
                atom_type=1
                q=1
                spin=part[0]
                rs=0
                r=part[1]
            else:
                atom_type=2
                q=0
                spin=part[0]
                rs=1.823572
                r=part[1]
            f.write("{:d}\t{:d}\t{:.6f}\t{:d}\t{:.6f}\t{:.6f}\t{:.6f}\t{:.6f}\n".format(
                i+1, atom_type, q, spin, rs, r[0], r[1], r[2]))
```

```python
write_lammps_data("test.lammps", a=17.7, N_h2=500)
```
